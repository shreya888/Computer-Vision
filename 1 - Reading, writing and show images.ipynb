#### Reading, writing and displaying images with OpenCV

Let's start by importing the OpenCV libary 

```python
# Import OpenCV
import cv2
# Import for colab
from google.colab.patches import cv2_imshow
# Import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
# Now let's import numpy as np
import numpy as np 
```

Let's now load our first image

```python
# Load an image using 'imread' specifying the path to image
im = cv2.imread('/Images/penguins.jpg')

# Our file 'penguins.jpg' is now loaded and stored in python 
# as a varaible we named 'im'

# To display our image, we use 'imshow' (here cv2_imshow because using Colab)
#cv2.imshow(im)
cv2_imshow(im)

# 'waitKey' allows us to input information when a image window is open
# By leaving it blank it just waits for anykey to be pressed before 
# continuing. By placing numbers (except 0), we can specify a delay for
# how long you keep the window open (time is in milliseconds here)
cv2.waitKey(100)

# This closes all open windows 
# Failure to place this will cause your program to hang
cv2.destroyAllWindows()
```

```python
# Load in Grayscale
gray_im = cv2.imread('/Images/penguins.jpg', 0)
cv2_imshow(gray_im)
```

```python
# Again taking the same image
im = cv2.imread('/Images/penguins.jpg')

# Plotting numpy arrays as images using plt.imshow
plt.imshow(im)
plt.show()
```

```python
# The ordering of color channels is BGR in OpenCV and RGB in matplotlib
# hence the colors are off
# This can be rectified using the cvtColor to convert to
# plot image in something other than OpenCV since it breaks in Colab
imrgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(imrgb)
plt.show()

# Or can use the numpy functions / indexing
imrgb = im[:,:,::-1]
plt.imshow(imrgb)
plt.show()
```

### Let's take a closer look at how images are stored

```python
# An image is represented as a numpy array in the format
# row (height) x column (width) x channels (3)
print(type(im))

# Shape gives the dimensions of the image array
print(im.shape)
```

```python
# Let's print each dimension of the image

print('Height of Image:', int(im.shape[0]), 'pixels')
print('Width of Image: ', int(im.shape[1]), 'pixels')
```

### How do we save images we edit in OpenCV?

```python
# Simply use 'imwrite' specificing the file name and the image to be saved
cv2.imwrite('/Images/output.jpg', im)
cv2.imwrite('/Images/output.png', im)
```
