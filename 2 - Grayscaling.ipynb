## Grayscaling

#### Grayscaling is process by which an image is converted from some other color scheme (RGB, MMYK, HSV etc) to shades of grey varying between complete black & complete white.

In OpenCV, many functions are customized to work on grayscale images (example: Canny edge detection function in OpenCV). This is done because it reduces dimensions, acting almost as a noise reduction and increasing processing time due to reduction in model complexity.

### Let convert our color image to greyscale

```python
import cv2

# Import for colab
from google.colab.patches import cv2_imshow

# Load our input image
im = cv2.imread('/Images/penguins.jpg')

# Original image
cv2_imshow(im)

cv2.waitKey()

# We use cvtColor, to convert to grayscale
gray_im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

# Image in Grayscale
cv2_imshow(gray_im)

cv2.waitKey()
```

```python
# Another method faster method is to load image in Grayscale
# The second argument of imread is a flag which specifies how image should be read
gray_im = cv2.imread('/Images/penguins.jpg', 0)

# Image in Grayscale
cv2_imshow(gray_im)

cv2.waitKey()
```
